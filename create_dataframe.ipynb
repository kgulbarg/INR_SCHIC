```python
# Create dataframe and put it in a .csv file with these columns:
# Target_label, cell_name, chr_name, bin_1, bin_2, count

data_root = "/home/kgulbarg/thesis/INR_SCHIC/Nagano_Data"
```

```python
import os
import pandas as pd
import csv
```

```python
for label in os.listdir(data_root):
    print(label)
```

```python
csv_path = "/home/kgulbarg/thesis/INR_SCHIC/Nagano.csv"
df = pd.DataFrame(columns = ["Target_label", "Cell_name", "Chr_name", "Bin_1", "Bin_2", "Count"])

print(df)
```

```python
# data runs into GBs. Writing directly to .csv
```

```python
headers = ["Target_label", "Cell_name", "Chr_name", "Bin_1", "Bin_2", "Count"]

with open(csv_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(headers)

    for target_label in os.listdir(data_root):
        target_label_dir = os.path.join(data_root, target_label)
        print(f"\n\nStarting type: {target_label}")

        no_of_cells = 0
        for cell_name in os.listdir(target_label_dir):
            cell_name_dir = os.path.join(target_label_dir, cell_name)
            # filter latent files. take only directories:
            if not os.path.isdir(cell_name_dir):
                continue
                
            no_of_chroms = 0
            for chr_name in os.listdir(cell_name_dir):
                chr_name_dir = os.path.join(cell_name_dir, chr_name)
                # filter any .ipynb checkpoints
                if not (os.path.isfile(chr_name_dir) and chr_name.startswith("chr") and chr_name.endswith(".txt")):
                    continue
                
                df = pd.read_csv(chr_name_dir, sep=r"\s+", engine="python",
                                 header=0, comment="#", 
                                 usecols=["bin1","bin2","count"])
        
                chr_stem = chr_name[:-4]
                
                for b1, b2, cnt in df.itertuples(index=False, name=None):
                    writer.writerow([target_label, cell_name, chr_stem, int(b1), int(b2), int(cnt)])

                # print(f"Done recording bins and contacts for chromosome: {chr_stem}")
                no_of_chroms += 1
            print(f"Done recording {no_of_chroms} chromosomes in cell: {cell_name}. Progress: {no_of_cells} cells of {target_label}.")
            no_of_cells += 1
        print(f"Done recording {no_of_cells} cells of type: {target_label}")
    print(f"Done making CSV file: {csv_path}")
```
